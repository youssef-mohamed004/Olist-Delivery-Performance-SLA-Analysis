## Geographic Performance Analysis 🌍

This analysis evaluates delivery SLA performance across different cities and states.  
The dataset was grouped by customer city and state to measure:

- **Late Rate** → Percentage of delayed orders.
- **Late Orders** → Total number of delayed deliveries.
- **Order Volume** → Total number of orders per location.

The objective is to identify geographic areas with poor delivery performance and detect regions that may require operational improvements or logistics optimization.

```python
Geographic_Performance["is_late"] = Geographic_Performance["Delivery_Status"]=="Late"
```

```python
city_geo = Geographic_Performance.groupby("customer_city")["is_late"].agg(late_rate="mean" ,late_orders= "sum" , order_NO = "count").reset_index()
city_geo = city_geo[city_geo["order_NO"] >= 100]
city_geo
```

# bottleneck

```python
bottleneck = seller_impact[
                            ["customer_id" 
                            ,"order_status"
                            ,"Delivery_Status"
                            ,"order_purchase_timestamp" 
                            , "order_approved_at" 
                            , "order_delivered_carrier_date" 
                            , "order_delivered_customer_date" 
                            , "order_estimated_delivery_date"
                            ,"shipping_limit_date"]
    ].copy()
```

```python
bottleneck = pd.merge(bottleneck , customer , on = "customer_id")
```

```python
bottleneck
```

```python
bottleneck["shipping_limit_date"] = pd.to_datetime(bottleneck["shipping_limit_date"] , errors = "coerce")
```

```python
bottleneck.info()
```

```python
bottleneck.isna().sum()
```

```python
bottleneck = bottleneck[bottleneck["order_approved_at"].notna()]
bottleneck = bottleneck[bottleneck["order_delivered_carrier_date"].notna()]
```

```python
bottleneck.isna().sum()
```

```python
bottleneck["processing_time"] = (bottleneck["order_approved_at"] - bottleneck["order_purchase_timestamp"]).dt.days

bottleneck["shipping_time"] = (bottleneck["order_delivered_customer_date"] - bottleneck["order_delivered_carrier_date"]).dt.days

bottleneck["seller_delay"] = (bottleneck["order_delivered_carrier_date"] - bottleneck["shipping_limit_date"]).dt.days

bottleneck["final_delay"] = (bottleneck["order_delivered_customer_date"] - bottleneck["order_estimated_delivery_date"]).dt.days
```

```python
x = bottleneck.groupby("Delivery_Status")[["processing_time" , "shipping_time" , "seller_delay" , "final_delay"]].mean()
```

```python
late_orders = bottleneck[bottleneck["Delivery_Status"] == "Late"]
total_late = len(late_orders)

shipping_caused  = len(late_orders[late_orders["shipping_time"] > bottleneck["shipping_time"].median()])
seller_caused    = len(late_orders[late_orders["seller_delay"] > 0])
processing_caused = len(late_orders[late_orders["processing_time"] > bottleneck["processing_time"].median()])

print(f"Shipping caused   = {shipping_caused/total_late*100:.2f}%")
print(f"Seller caused     = {seller_caused/total_late*100:.2f}%")
print(f"Processing caused = {processing_caused/total_late*100:.2f}%")
```
